```python
import pandas as pd
import numpy as np

Data_Listings = pd.read_csv('listings.csv')
Data_Listings.head()
```

```python
Data_Listings.id = Data_Listings.id.astype('Int64')
Data_Listings.host_id = Data_Listings.host_id.astype('Int64')
Data_Listings.describe()
```

```python
#сохраняем колонки, в которых 16 не пустых значений (т.к. всего 18, а 2 их них пустые всегда)
Data_Listings_Drop = Data_Listings.dropna(thresh=16)
Data_Listings_Drop.describe()
```

```python
Data_Listings_Drop
```

```python
#удаляем пустые колонки
Data_Listings_Drop = Data_Listings_Drop.dropna(axis=1)
Data_Listings_Drop.describe()
```

```python
Data_Listings_Drop.price.agg(['mean', 'median'])
```

```python
Data_Listings_Drop.groupby(['room_type']).agg({'price':['median','mean','min','max']})
```

```python
Data_Listings_Drop.groupby(['room_type']).agg({'last_review': ['max','min']})
```

```python
Data_Listings_Drop.groupby(['neighbourhood']).agg({'last_review': ['max','min']})
```

```python
Data_Listings_Drop.groupby(['minimum_nights']).agg({'price': ['max','min','mean','median'], 'reviews_per_month':['mean'] })
```
